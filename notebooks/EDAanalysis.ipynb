# Resume Data Analysis and Visualization

## Overview
This Jupyter Notebook performs basic exploratory data analysis (EDA) on a sample resume dataset. The dataset includes information about individuals' experience, education, skills, and location. 

## Steps Performed:
1. **Creating a Sample Dataset**  
   - The dataset consists of fictional resume data with columns: `name`, `experience`, `education`, `skills`, and `location`.

2. **Displaying Dataset Information**  
   - The `data.info()` function is used to check the structure and types of data in the dataset.

3. **Displaying the First Few Rows**  
   - The first few rows of the dataset are printed using `data.head()` to get an overview of the data.

4. **Visualizing the Experience Distribution**  
   - A histogram is plotted using `seaborn` to show the distribution of years of experience among individuals.  
   - `sns.histplot(data['experience'], kde=True)` is used to generate the plot, where `kde=True` overlays a probability density function.

5. **Checking for Missing Values**  
   - `data.isnull().sum()` is used to check for missing values in each column, ensuring data completeness.

## Expected Outcome:
- A summary of the dataset structure.
- A table showing sample data.
- A histogram illustrating the distribution of experience levels.
- A count of missing values per column.

This analysis provides insights into the dataset, which can later be used for preprocessing and machine learning applications.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a sample dataset
data = pd.DataFrame({
    'name': ['Alice Johnson', 'Bob Smith', 'Charlie Brown', 'David Lee', 'Emma Wilson'],
    'experience': [3, 7, 5, 10, 2],
    'education': ['Bachelors', 'Masters', 'PhD', 'Bachelors', 'Masters'],
    'skills': ['Python, SQL', 'Machine Learning, Python', 'Data Science, R', 'Web Development, JavaScript', 'Cloud Computing, AWS'],
    'location': ['New York', 'San Francisco', 'Chicago', 'Los Angeles', 'Seattle']
})

# Display basic information
print(data.info())

# Display first few rows
print(data.head())

# Visualizing the distribution of job experiences
sns.histplot(data['experience'], kde=True)
plt.title('Experience Distribution')
plt.xlabel('Years of Experience')
plt.ylabel('Count')
plt.show()

# Check for missing values
print("Missing values in each column:\n", data.isnull().sum())
```
